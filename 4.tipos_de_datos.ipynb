## Tipos de datos
-----------

- Categóricos:
  Pueden ser divididos en categorías 
    - Ordinales.- Cuando existe una relación de orden entre las categorías.
    - Nominales.- No tienen relaciones categóricas.


- Numéricos:
  Pueden ser agrupados en intervalos

    - Discretos.- Son enteros. (*int*)
    - Continuos.- Pueden tener decimales. (*float* y *doubles*)
