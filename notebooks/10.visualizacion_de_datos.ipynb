```python
# %%
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# %%
```

```python
iris = sns.load_dataset("iris")
iris.head(5)
```

#### **Diagrama de Dispersi贸n**
Documentaci贸n => [Scatterplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```python
# iris_sp = 
plt.grid(color='black', linestyle='dotted')
sns.scatterplot(data=iris, x='sepal_length' , y='petal_length', hue='species', palette='colorblind', style='species')
plt.ylabel('Longitud de petalo (mm)')
plt.xlabel('Longitud de sepalo (mm)')
plt.show()
```

Documentaci贸n => [Jointplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```python
# iris_sp = 
# plt.grid(color='black', linestyle='dotted')
sns.jointplot(data=iris, x='sepal_length' , y='petal_length', hue='species')
plt.ylabel('Longitud de petalo (mm)')
plt.xlabel('Longitud de sepalo (mm)')
plt.show()
```

#### **Diagrama de Distribuci贸n**
Documentaci贸n => [Boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

```python
sns.boxplot(data=iris, x='species' , y='sepal_length', palette='colorblind')#, hue='species')
plt.show()
```

#### **Diagrama de Distribuci贸n**
Documentaci贸n => [Barplot](https://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
sns.barplot(data=iris, x='species' , y='sepal_length')#, hue='species')
plt.show()
```
